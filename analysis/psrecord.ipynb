```python
STANDARD_FEDNN_PSRECORD_FILE = "/home/lukas/Dokumente/Bildung und Studium/Universit√§t Leipzig/Masterarbeit/Experiment/Standard-FedNN/1/flwr_client0_PID2484830.txt"
FHE_FEDNN_PSRECORD_FILE = "/home/lukas/Downloads/1/flwr_client0_PID3828851.txt"
```

```python
import pandas as pd

df_standard_fnn = pd.read_csv(
    STANDARD_FEDNN_PSRECORD_FILE, delim_whitespace=True, skiprows=1,
    names=["Elapsed Time (s)", "CPU (%)", "Real Memory (MB)", "Virtual Memory (MB)"]
)

print(df_standard_fnn.head())
```

```python
print(df_standard_fnn.describe())
```

```python
df_fhe_fnn = pd.read_csv(
    FHE_FEDNN_PSRECORD_FILE, delim_whitespace=True, skiprows=1,
    names=["Elapsed Time (s)", "CPU (%)", "Real Memory (MB)", "Virtual Memory (MB)"]
)
print(df_fhe_fnn.describe())
```

```python
import matplotlib.pyplot as plt

# Plot Real Memory Usage
plt.figure(figsize=(10, 5))
plt.plot(df_standard_fnn["Elapsed Time (s)"], df_standard_fnn["Real Memory (MB)"], label="Process 1", color="blue")
plt.plot(df_fhe_fnn["Elapsed Time (s)"], df_fhe_fnn["Real Memory (MB)"], label="Process 2", color="red")
plt.xlabel("Time (s)")
plt.ylabel("Real Memory (MB)")
plt.title("Memory Usage Over Time")
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10, 5))
plt.plot(df_fhe_fnn["Elapsed Time (s)"], df_fhe_fnn["CPU (%)"], label="FHE FNN", color="red")
plt.plot(df_standard_fnn["Elapsed Time (s)"], df_standard_fnn["CPU (%)"], label="Standard FNN", color="blue")
plt.xlabel("Time (s)")
plt.ylabel("CPU (%)")
plt.title("CPU Usage Over Time")
plt.legend()
plt.show()
```
